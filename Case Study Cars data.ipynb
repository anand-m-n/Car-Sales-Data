Case study on Presentation of data
Do the following operations on ‘cars_data’ dataset:
1. Read the dataset to the python environment.
2. Check for the null values present in the dataset.
3. Plot a bar graph of male vs female buyers participated in the sales.
4. Find the top 5 cars based on their sales price.
5. Find the least 5 cars based on their Resell price.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read the dataset to the python environment 

```python
cars_data = pd.read_csv('cars_data.csv')
cars_data
```

### Check for the null values present in the dataset. 

```python
cars_data.isnull().values.any()
```

### Plot a bar graph of male vs female buyers participated in the sales.

```python
plt.figure(figsize=(6,6))
graph= sns.countplot( x='Buyer Gender', data= cars_data, hue = 'Buyer Gender')
graph.get_legend().remove()

i=0

for p in graph.patches:
        height = p.get_height()
        graph.text(p.get_x()+p.get_width()/2., height + 0.5,
            height, ha="center")
        i += 1
```

5052 female and 4948 male buyers participated in the sales

### Find the top 5 cars based on their sales price. 

```python
top_sale = cars_data.sort_values('Sale Price', ascending= False).head(5)      
top_sale[['Make','Model', 'Sale Price']].style.hide_index()
```

### Find the least 5 cars based on their Resell price.

```python
low_resale = cars_data.sort_values('Resell Price', ascending= True).head(5)      
low_resale[['Make','Model','Resell Price']].style.hide_index()
```
